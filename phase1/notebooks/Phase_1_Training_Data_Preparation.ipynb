```python
codes_to_find = list(tickers["code"])
gics_sub_codes = list(gics_mapping["Sub_Industry_id"])
gics_group_ind_id = list(gics_mapping["Industry_Group_id"])
gics_group_ind_name = list(gics_mapping["Industry_Group_name"])

final_df = pd.DataFrame(columns=['ticker', 'Industry_group_code','Industry_group_name'])

for tick, code in zip(list(tickers["ticker"]), codes_to_find):
  if code in gics_sub_codes:
    index = gics_sub_codes.index(code)
    new_row = {'ticker':tick, 'Industry_group_code':gics_group_ind_id[index], 'Industry_group_name':gics_group_ind_name[index]}
    final_df = final_df.append(new_row, ignore_index=True)

display(final_df)
final_df.to_csv("data_for_model.csv", index=False)
```
